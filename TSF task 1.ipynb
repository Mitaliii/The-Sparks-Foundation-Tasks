# Data Visualization 

```python
data.hist(bins=50 ,figsize=(20,15))
plt.show()
```

## From the histogram Graph we can say that our data is not normal.

```python
sns.relplot(data=data,x='Profit',y='Sub-Category',hue='Category')
```

##### Above plot shows the profit and loss of each and every subcategories.Here from graph we can visualize that "binders" sub-category has suffered the highest amount of loss and also profit amongst all other sub-Categories (For now we can't say that what is the reason it may be because of discounts given on binders subcategory).

##### Next,"Copiers" Sub-category has gain highest amount of profit with no loss.There are other sub-categories too who are not faced any kind of losses but their profit margins are also low.

##### Next,Suffering from highest loss is machines.

```python
sns.relplot(data=data,x='Sales',y='Sub-Category',hue='Category')
```

###### Above plot shows the sales of each and every subcategories.Here from graph we can visualize that "Machines" sub-category has the highest sales amongst all other sub-Categories as an exception.
##### Next,"Copiers" Sub-category has highest sales .
##### "Fasteners","Envelopes","Paper","Labels" and  "Art" have lower Sales as compared to other sub-category

```python
figsize=(15,10)
sns.pairplot(data,hue='Sub-Category')
```

##### From the above plot we can say that Our Data is not Normal and it has some amount of outliers too

```python
sns.barplot(x="Region", y="Profit", hue="Category", data=data)
```

##### Above graph interprets that "Technology" category has the highest profit as compared to "Furniture" and "Office Supplies" also the "East" region have highest profit in every category. 

```python
sns.barplot(x="Region", y="Sales", hue="Category", data=data)
```

##### Above graph interprets that "Technology" category has the highest Sales as compared to "Furniture" and "Office Supplies" also the "East" region have highest sales in every category. 

```python
### sum,mean,min,max,count median,standard deviation,Variance of each states of Profit
data.groupby("State").Profit.agg(["sum","mean","min","max","count","median","std","var"])
```

##### From above Summary we can say that highest profit or loss sales has been done from which States,what is the varient profit is it high or low.What is the median Profit ,what is the Mean profit ,actually Mean will Give an Unbiased Result Compared to median Because Mean considered whole data in Calculation while Median is not.

```python
import seaborn as sns
sns.barplot(x="Category", y="Sales", hue="Ship Mode", data=data)
```

##### Above graph interprets that "Technology" category has the highest Sales in the same day(ship mode) as compared to "Furniture" and "Office Supplies" whereas, ship mode in same day has the lowest sales in office supplies category  

```python
sns.barplot(x="Category", y="Profit", hue="Region", data=data)
```

##### Above graph interprets that "Technology" category has the highest Profit in the East region whereas, "Office supplies" and "Furniture" category have loss in the central region

```python
data.groupby('Ship Mode').size().plot(kind='pie',autopct='%1.2f',colors={'cyan','tomato','violet','yellow'})
```

##### Above pie chart shows total percentage of each ship mode

```python
fig = plt.gcf()
fig.set_size_inches(16,5)
sns.scatterplot(x=data['Sub-Category'],y=data['Sales'],hue=data['Profit'])
```

##### Above plot is the sales of each sub-category clustered with respect to profit and loss it sub-category made

```python
data.groupby('Region').size().plot(kind='pie',autopct='%.2f',colors={'cyan','tomato','violet','yellow'})
```

##### Above pie chart shows total percentage of each region

```python
data.groupby('Segment').size().plot(kind='pie',autopct='%.1f',colors={'cyan','pink','gold'})
```

##### Above pie chart shows total percentage of each segment

```python
g = sns.pairplot(data,hue="Region")
g.map(sns.scatterplot)
g.add_legend()
```

```python
g = sns.PairGrid(data,hue="Segment")
g.map(sns.scatterplot)
g.add_legend()
```
